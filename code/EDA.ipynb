```python
def time_of_day(hour):
    if 6 <= hour < 12:
        return '6-12'
    elif 12 <= hour < 18:
        return '12-18'
    elif 18 <= hour < 24:
        return '18-24'
    else:
        return '0-6'
    
rainbow_palette = sns.color_palette("rainbow", 7)

cancellation_rate['TimeOfDay'] = cancellation_rate['Hour'].apply(time_of_day)


plt.figure(figsize=(12, 8))
sns.lineplot(data=cancellation_rate, x='TimeOfDay', y='CancellationRate', hue='DayOfWeek', palette=rainbow_palette, marker="o")
plt.xlabel("Time of Day", fontsize=14)  # Adjust font size for x-axis label
plt.ylabel("Cancellation Rate", fontsize=14)  # Adjust font size for y-axis label
plt.title("Cancellation Rate by Time of Day", fontsize=16)  # Adjust font size for title

plt.legend(title="Day of Week", loc='upper right', fontsize=12, title_fontsize=14)
plt.show()
```

```python
features = ['Hour','OriginAirportID','Distance','HourlyPrecipitation_originairport', 'HourlyWindSpeed_originairport', 'HourlyVisibility_originairport','HourlyVisibility_destairport','HourlyWindSpeed_destairport','HourlyPrecipitation_destairport']
target = 'DepDelayMinutes'
df_not_cancelled = df[df['Cancelled'] == 0]
# Drop rows with missing values in the selected columns
df_not_cancelled = df_not_cancelled.dropna(subset=features + [target])
df_not_cancelled['HourlyWindSpeed_originairport'] = pd.to_numeric(
    df_not_cancelled['HourlyWindSpeed_originairport'], errors='coerce'
)
df_not_cancelled['HourlyPrecipitation_originairport'] = pd.to_numeric(
    df_not_cancelled['HourlyPrecipitation_originairport'], errors='coerce'
)
df_not_cancelled['HourlyVisibility_originairport'] = pd.to_numeric(
    df_not_cancelled['HourlyVisibility_originairport'], errors='coerce'
)
df_not_cancelled['HourlyVisibility_destairport'] = pd.to_numeric(
    df_not_cancelled['HourlyVisibility_destairport'], errors='coerce'
)
df_not_cancelled['HourlyWindSpeed_destairport'] = pd.to_numeric(
    df_not_cancelled['HourlyWindSpeed_destairport'], errors='coerce'
)
df_not_cancelled['HourlyPrecipitation_destairport'] = pd.to_numeric(
    df_not_cancelled['HourlyPrecipitation_destairport'], errors='coerce'
)

df_not_cancelled = df_not_cancelled.dropna(subset=['HourlyWindSpeed_originairport','HourlyPrecipitation_destairport','HourlyWindSpeed_destairport','HourlyVisibility_destairport','HourlyVisibility_originairport','HourlyPrecipitation_originairport'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
skewness = df_not_cancelled['DepDelayMinutes'].skew()
print(f"Skewness: {skewness}")
import matplotlib.pyplot as plt

# Plot histogram
plt.hist(df_not_cancelled['DepDelayMinutes'], bins=30, edgecolor='k')
plt.title('Histogram of Delay Minutes')
plt.xlabel('Delay Minutes')
plt.ylabel('Frequency')
plt.show()
#Negative skewness (less than 0) means the distribution is skewed to the left (long tail on the left).
```

```python
import numpy as np
df_not_cancelled['Log_DepDelayMinutes']= np.log1p(df_not_cancelled['DepDelayMinutes'])  # log1p to handle zero values
transformed_skewness = df_not_cancelled['Log_DepDelayMinutes'].skew()
print(f"Transformed Skewness: {transformed_skewness}")
```

```python
target = 'Log_DepDelayMinutes'
df_not_cancelled = df_not_cancelled.dropna(subset='Log_DepDelayMinutes')
X = df_not_cancelled[features]
y = df_not_cancelled[target]

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()

# Train the model
linear_model.fit(X_train, y_train)
```

```python
y_pred = linear_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
