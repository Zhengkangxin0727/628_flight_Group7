```python
data['HourlyPresentWeatherType_originairport'].value_counts(normalize=True).head(5)
```

```python
import numpy as np
def process_HourlyPresentWeatherType(data, columns):
    for column in columns:
        if column in data.columns:
            data[column] = data[column].apply(
                lambda x: str(x).split('|')[-1].strip()[:2]
                if isinstance(x, str) and len(str(x).split('|')) > 1
                else ''
            )
            data[column] = data[column].replace({'*': np.nan, '': np.nan})
            data[column] = data[column].fillna('NORM')

    return data
columns = ['HourlyPresentWeatherType_originairport', 'HourlyPresentWeatherType_destairport']
data = process_HourlyPresentWeatherType(data, columns)
print(data[columns].head())
```

##HourlySkyConditions_originairport

```python
data[['HourlySkyConditions_originairport']].head()
```

```python
import pandas as pd
import numpy as np
import re
def get_last_sky_condition(row):
    if isinstance(row, str):
        conditions = row.split(' ')
        last_condition = conditions[-2] if len(conditions) >= 2 else conditions[-1]
        sky_status = last_condition.split(':')[0]
        cloud_height = last_condition.split(':')[1] if ':' in last_condition else np.nan
        if isinstance(cloud_height, str):
            cloud_height = re.sub(r'[^\d.]+', '', cloud_height)

        return pd.Series([sky_status, cloud_height])
    else:
        return pd.Series([np.nan, np.nan])
def process_sky_conditions(data, columns, delete_original=False):
    for column in columns:
        if column in data.columns:
            sky_condition_col = f"SkyCondition_{column}"
            cloud_height_col = f"CloudHeight_{column}"
            data[[sky_condition_col, cloud_height_col]] = data[column].apply(get_last_sky_condition)
            data[cloud_height_col] = data[cloud_height_col].astype(float).apply(np.floor).astype("Int64")

            if delete_original:
                data.drop(columns=[column], inplace=True)

    return data

columns = ['HourlySkyConditions_originairport', 'HourlySkyConditions_destairport']
data = process_sky_conditions(data, columns, delete_original=True)

print(data[['SkyCondition_HourlySkyConditions_originairport', 'CloudHeight_HourlySkyConditions_originairport']].head())
print(data[['SkyCondition_HourlySkyConditions_destairport', 'CloudHeight_HourlySkyConditions_destairport']].head())
```

```python
import pandas as pd

def split_time_column(data, column_name):

    data[column_name] = pd.to_datetime(data[column_name], errors='coerce')

    data[f'{column_name}_hour'] = data[column_name].dt.hour
    data[f'{column_name}_min'] = data[column_name].dt.minute

    data = data.drop(column_name, axis=1)

    return data

time_columns = ['CRSDepTime_CST', 'DepTime_CST', 'CRSArrTime_CST', 'ArrTime_CST']
for col in time_columns:
    data = split_time_column(data, col)

print(data.head())
```
